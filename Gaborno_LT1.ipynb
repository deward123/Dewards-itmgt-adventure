### Turning Tables




```python
i=1

x = int(input("Number of rows:"))
y = 2

while (i<x):
    j=0
    while(j<y):
        if i%2 == 0:
            print(i,end=" ")
        
        else:
            print(i*i,end=" ")
        j+= 1
    print()
    i += 1
    
```

### 4. Go forth and Multiply



```python
def Multtable(t):
    for i in range(1,t+1):
        for j in range(1,t+1):
            print(i*j, end="\t")
        print()
```

```python
Multtable(3)
```
